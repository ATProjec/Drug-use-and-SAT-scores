### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
#a variable that turns the sum of two numbers in to a string
number = str(99*91)

#a function that takes an argument (number) and returns number in a reversed order
def isPalindrome(number):
    return number == number[::-1]

#an if condition that will print a confirmation statement if the given number is palindromic
if isPalindrome(number) is True:
    print('Your number is a palindromic number')
else:
    print('Your number is not a palindromic number')
```

I initiated a variable called number.

Then I created a function called isPalindrome() that takes a argument called number.
This function returns a True value if our number variable equates to the reverse of number which is denoted by [::-1].

Once the function has been created, I called upon an if statement where if our function is true, it will print "Your number is a palindromic number'. Otherwise, any other condition (false) will be met with "Your number is not a palindromic number".


### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
#a variable assigned to 0, this variable will be used to contain a running total for our code
total = 0

#a for loop that limits the iterations to 2000
for i in range(2000):
    #following statements will only run if variable i is larger than 1
    if i>1:
        #another for loop that starts at 2 and is limited to variable i
        for j in range(2,i):
            #variable j will act as the same number as variable i in order to find modulus 0
            if(i % j==0):
                #if it is mod 0 then this if statement will break (stop)
                break
        #once the if loop above breaks, this else statement will run and allow variable i to be added to total
        else:
            total += i
print(total)
```

I created a variable called total, a nested for loop and if statement block of code. it can be difficult to understand at first but bear with me.

I generated a range of 2000, i variable will behave as a counter for this for loop.

Within this for loop, if i is greater than 1 then it can proceed with the next nested line which is a for loop.

A variable j is created where starting at a range of 2 to i (which is limited to 2000). Every time this for loop statement loops, it will add i to total unless i modolus j has a remainder of 0.

### Challenge 4: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def compress(string):

    new = ""

    count = 1

    #Adds first character of string to variable new
    new += string[0]

    #loops through the range of the string length minus 1
    for i in range(len(string)-1):
        #if the first letter is the same to the second letter then the count will increase by 1
        if(string[i] == string[i+1]):
            count += 1
        else:
            if(count > 1):
                #the count will be added to the new string as a string
                new += str(count)
            #otherwise, the letter and 1 will be added to the new string
            new += string[i+1]
            #count is reset
            count = 1
    if(count > 1):
        new += str(count)
    return new
compress("aabccccaa")
```
